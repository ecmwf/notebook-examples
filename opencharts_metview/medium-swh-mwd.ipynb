## Retrieve Data
This product takes in input 2 parameters :

* [Significant height of combined wind waves and swell](https://codes.ecmwf.int/grib/param-db/140229)
* [Mean wave direction](https://codes.ecmwf.int/grib/param-db/140230)

In this example, we will use: 
- [**ecmwf.opendata**](https://github.com/ecmwf/ecmwf-opendata) Client to download the data
- [**Metview**](https://metview.readthedocs.io/en/latest/) library to read, process and plot the data

```python
import metview as mv
from ecmwf.opendata import Client
```

```python
client = Client("ecmwf", beta=True)
```

```python
parameters = ['swh', 'mwd']
filename = 'medium-swh-mwd.grib'
filename
```

Setting date to 0 will download today's data. 
Removing date and time altogether from the request will download the latest data.  
Try commenting out date and time to download latest forecast!

```python
client.retrieve(
    date=0,
    time=0,
    step=12,
    stream="wave",
    type="fc",
    levtype="sfc",
    param=parameters,
    target=filename
)
```

## Reading and processing the data
Now we can use **ecmwf.data** to read the files.

```python
data = mv.read(filename)
```

The **describe()** function will give us the overview of the dataset.  

```python
data.describe()
```

And an overview of one parameter, where we can see more information, such as units or type of level.

```python
data.describe('swh')
```

We can use **ls()** function to list all the fields in the file we downloaded.

```python
data.ls()
```

The grib file contains all the parameters, and we will use the **select()** function to filter what we need.  
Feel free to use **describe()** to inspect the other filtered fields.

```python
swh = data.select(shortName= 'swh')
mwd = data.select(shortName= 'mwd')

swh.describe()
```

We need to create a wind vector field for the visualisation, as well.

```python
wave_vector = mv.grib_vectors(type='polar_field', intensity = swh, direction = mwd)
```

And finally, we can plot the data on the map. 

```python
# define coastlines
coast = mv.mcoast(
    map_coastline_colour="charcoal",
    map_coastline_resolution="medium",
    map_coastline_land_shade="on",
    map_coastline_land_shade_colour="cream",
    map_coastline_sea_shade="off",
    map_boundaries="on",
    map_boundaries_colour= "charcoal",
    map_boundaries_thickness = 1,
    map_disputed_boundaries = "off",
    map_grid_colour="tan",
    map_label_height=0.35,
)

# define view
view = mv.geoview(
    area_mode="name",
    area_name="north_atlantic",
    subpage_clipping="on",
    coastlines=coast
)

#define styles
swh_shade = mv.mcont(legend= "on",
                contour_automatics_settings = "style_name",
                contour_style_name = "sh_all_f0t20lst")

wave_arrows = mv.mwind(
    wind_thinning_factor=5, wind_arrow_colour="black"
)

title = mv.mtext(
    text_lines=["Significant wave height and mean direction", 
                "START TIME: <grib_info key='base-date' format='%a %d %B %Y %H' where='shortName=mwd'/> ",
                "VALID TIME: <grib_info key='valid-date' format='%a %d %B %Y %H' where='shortName=swh'/>, STEP: <grib_info key='step' where='shortName=swh' />"],
    text_font_size=0.4,
    text_colour         = 'charcoal')

ecmwf_text = mv.mtext(    
    text_lines          = [" European Centre for Medium-Range Weather Forecasts (ECMWF)",
                           "Source: www.ecmwf.int Licence: CC-BY-4.0 and ECMWF Terms of Use",
                            "https://apps.ecmwf.int/datasets/licences/general/"],
    text_justification  = 'center',
    text_font_size      = 0.3,
    text_mode           = "positional",
    text_box_x_position = 6.,
    text_box_y_position = -0.2,
    text_box_x_length   = 8,
    text_box_y_length   = 2,
    text_colour         = 'charcoal')

# generate plot
mv.setoutput('jupyter', plot_widget=False)
mv.plot(view, swh, swh_shade, wave_vector, wave_arrows, title, ecmwf_text)
```

To generate the png file you can run the following cell.

```python
png = mv.png_output(
      output_name = "medium-swh-mwd",   # specify relative or full path
      output_title = "medium-swh-mwd",    # title used by a viewer
      output_width = 1000,                 # set width in pixels
)
mv.setoutput(png)
mv.plot(view, swh, swh_shade, wave_vector, wave_arrows, title, ecmwf_text)
```

Note that plot produced using open data dataset will slightly differ from one from Open Charts. This is due to different resolution of the data.  
Open data is on 0.25x0.25 resolution, while high resolution data is 0.1x0.1 grid.
